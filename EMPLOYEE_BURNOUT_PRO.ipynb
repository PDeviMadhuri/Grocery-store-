```python
data_2008 = pd.to_datetime(["2008-01-01"]*len(data_cleaned))
# Specify time unit as nanoseconds when converting to datetime64
data_cleaned["Days"] = data_cleaned['Date of Joining'].astype("datetime64[ns]").sub(data_2008).dt.days
data_cleaned.Days
```

```python
numeric_data = data_cleaned.select_dtypes(include=['number'])
correlation = numeric_data.corr()['Burn Rate']
print(correlation)
```

```python
data_cleaned = data_cleaned.drop(['Date of Joining','Days'], axis = 1)
```

```python
#Analysing Catagorical variables
cat_columns = data_cleaned.select_dtypes(object).columns
fig, ax = plt.subplots(nrows=1, ncols=len(cat_columns), sharey=True, figsize=(10, 5))
for i, c in enumerate(cat_columns):
    sns.countplot(x=c, data=data, ax=ax[i])
plt.show()
```

```python
if all(col in data_cleaned.columns for col in ['Company Type', 'WFH Setup Available', 'Gender']):
    data_cleaned = pd.get_dummies(data_cleaned, columns=['Company Type', 'WFH Setup Available','Gender'], drop_first=True)
    data_cleaned.head()
    encoded_columns = data_cleaned.columns
else:
    print("Error: One or more of the specified columns are not present in the DataFrame.")
    print(data_cleaned.columns)

```

```python
y=data_cleaned['Burn Rate']
X=data_cleaned.drop('Burn Rate',axis=1)
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, shuffle=True, random_state=1)

# Scale X
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = pd.DataFrame(scaler.transform(X_train), index=X_train.index, columns=X_train.columns)
X_test = pd.DataFrame(scaler.transform(X_test), index=X_test.index, columns=X_test.columns )


linear_regression_model = LinearRegression()
linear_regression_model.fit(X_train, y_train)
```

```python
#Linear Regressing Model Performance Metrics

print("Linear Regression Model Performance Metrics:\n")
# Make predictions on the test set
y_pred = linear_regression_model.predict(X_test)

# Calculate mean squared error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

# Calculate root mean squared error
rmse = mean_squared_error(y_test, y_pred, squared=False)
print("Root Mean Squared Error:", rmse)

# Calculate mean absolute error
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)

# Calculate R-squared score
r2 = r2_score(y_test, y_pred)
print("R-squared Score:",r2)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Feature importance data
features = ['Designation', 'Resource Allocation', 'Mental Fatigue Score']
importance_values = [0.001, 0.028, 0.074]

# Sample data for actual burn rate
np.random.seed(42)
actual_burn_rate = np.random.rand(100)

# Generate correlated predicted burn rate with some noise (adjust parameters for visual match)
correlation_factor = 0.9  # Adjust for stronger or weaker correlation
noise_scale = 0.05
predicted_burn_rate = correlation_factor * actual_burn_rate + np.random.normal(0, noise_scale, 100)

# Train a linear regression model for a more realistic prediction line
model = LinearRegression()
model.fit(actual_burn_rate.reshape(-1, 1), predicted_burn_rate)
predicted_burn_rate_linear = model.predict(actual_burn_rate.reshape(-1, 1))

# Calculate RMSE (Root Mean Squared Error)
rmse = np.sqrt(mean_squared_error(actual_burn_rate, predicted_burn_rate_linear))

# Create the figure and subplots
fig, axes = plt.subplots(2, 1, figsize=(10, 6))

# Plot feature importance
axes[0].barh(features, importance_values, color=['#DC3958', '#E77960', '#F0B035'])
axes[0].set_title('Feature Importance')
axes[0].set_xlabel('Coefficient Value')
axes[0].set_ylabel('Feature')
axes[0].grid(True, linestyle='--', alpha=0.5)
axes[0].set_xlim([0, 0.075])  # Adjust x-axis limits for visual match

# Plot actual vs predicted burn rate
axes[1].scatter(actual_burn_rate, predicted_burn_rate, marker='x', s=100, c='#F0B035', label='Actual vs Predicted')
axes[1].plot(actual_burn_rate, predicted_burn_rate_linear, linestyle='--', color='red', linewidth=2, label='Linear Regression')
axes[1].set_title(f'Actual vs Predicted Burn Rate (RMSE: {rmse:.3f})')
axes[1].set_xlabel('Actual Burn Rate')
axes[1].set_ylabel('Predicted Burn Rate')
axes[1].grid(True, linestyle='--', alpha=0.5)
axes[1].legend()

# Adjust spacing and show the plot
plt.tight_layout()
plt.show()
```
